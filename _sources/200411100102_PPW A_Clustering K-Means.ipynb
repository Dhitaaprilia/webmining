```python
from sklearn.decomposition import LatentDirichletAllocation
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/PPW/Tugas/term Frequensi.csv')
df
```

```python
X = df.drop('Dokumen', axis=1)
X
```

```python
k = 3
alpha = 0.1
beta = 0.2

lda = LatentDirichletAllocation(n_components=k, doc_topic_prior=alpha, topic_word_prior=beta)
proporsi_topik_dokumen = lda.fit_transform(X)
```

```python
dokumen = df['Dokumen']
output_proporsi_TD = pd.DataFrame(proporsi_topik_dokumen, columns=['Topik 1', 'Topik 2', 'Topik 3'])
# Mendefinisikan peraturan pemetaan label
def map_to_label(probabilities):
    # Misalnya, kita mengasumsikan bahwa label "RPL" adalah label topik pertama jika probabilitasnya paling tinggi,
    # dan label "KOMPUTASI" adalah label topik kedua jika probabilitasnya paling tinggi.
    if probabilities['Topik 1'] > probabilities['Topik 2']:
        return 'RPL'
    else:
        return 'KK'
output_proporsi_TD['Label'] = output_proporsi_TD.apply(map_to_label, axis=1)

output_proporsi_TD.insert(0,'Dokumen', dokumen)
output_proporsi_TD
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Baca data
data_x = pd.read_csv('https://raw.githubusercontent.com/Feb11F/dataset/main/Term%20Frequensi%20Berlabel%20Final.csv')
data_x = data_x.dropna(subset=['Dokumen'])  # Menghapus baris yang memiliki NaN di kolom 'Dokumen'

# Ubah kelas A menjadi 0 dan kelas B menjadi 1
kelas_dataset_binary = [0 if kelas == 'RPL' else 1 for kelas in data_x['Label']]
data_x['Label'] = kelas_dataset_binary

# Bagi data menjadi data pelatihan dan data pengujian
X = data_x['Dokumen']
label = data_x['Label']
X_train, X_test, y_train, y_test = train_test_split(X, label, test_size=0.2, random_state=42)

# Vektorisasi teks menggunakan TF-IDF
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# Latih model Naive Bayes
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train_tfidf, y_train)

# Latih model LDA
k = 3
alpha = 0.1
beta = 0.2
lda = LatentDirichletAllocation(n_components=k, doc_topic_prior=alpha, topic_word_prior=beta)
proporsi_topik_dokumen = lda.fit_transform(X_train_tfidf)

# Input dokumen yang ingin diklasifikasikan
input_dokumen = "penjadwalan kuliah perguruan kompleks permasalahan variabel dipertimbangkan jadwal optimal penyusunan jadwal kuliah selayaknya dicoba menemukan penjadwalan terbaik dibutuhkan metode optimasi diterapkan penjadwalan mata kuliah metode menyelesaikan permasalahan algoritma genetika tugas bertujuan jadwal matakuliah otomatis hasilnya optimal hasil penelitian algoritma genetika jadwal optimal dibuktikan nilai fitness memiliki nilai otomatis jadwal dibentuk optimal semoga aplikasi diharapkan mempermudah didalam pembuatan penjadwalan matakuliah mudah cepat kunci algoritma genetika optimasi penjadwalan"
input_vector = tfidf_vectorizer.transform([input_dokumen])

# Prediksi kelas menggunakan model Naive Bayes
kelas_prediksi = nb_classifier.predict(input_vector)[0]

# Prediksi proporsi topik menggunakan model LDA
proporsi_topik = lda.transform(input_vector)[0]

# Tampilkan hasil prediksi
print("Prediksi Kelas (Naive Bayes):", "RPL" if kelas_prediksi == 0 else "KK")
print("Proporsi Topik (LDA):", proporsi_topik)
```

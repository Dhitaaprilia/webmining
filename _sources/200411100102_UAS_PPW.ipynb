## Cek Null Data

## Cleaning

```python
def cleaning(text):
  text = re.sub(r'[^a-zA-Z\s]', '', text).strip()
  return text

df['Cleaning'] = df['Artikel'].apply(cleaning)
df['Cleaning']
```

## Tokenizing

```python
def tokenizer(text):
  text = text.lower()
  return word_tokenize(text)

df['Tokenizing'] = df['Cleaning'].apply(tokenizer)
df['Tokenizing']
```

## Stopword Removal

```python
corpus = stopwords.words('indonesian')

def stopwordText(words):
 return [word for word in words if word not in corpus]

df['Stopword Removal'] = df['Tokenizing'].apply(stopwordText)

# Gabungkan kembali token menjadi kalimat utuh
df['Full Text'] = df['Stopword Removal'].apply(lambda x: ' '.join(x))
df['Full Text']
```

## TF-IDF

```python
def tfidf(dokumen, category):
  vectorizer = TfidfVectorizer()
  x = vectorizer.fit_transform(dokumen).toarray()
  terms = vectorizer.get_feature_names_out()

  final_tfidf = pd.DataFrame(x, columns=terms)
  final_tfidf.insert(0, 'Dokumen', dokumen)
  final_tfidf.insert(len(final_tfidf.columns),'Category', category)

  return (vectorizer, final_tfidf)

tfidf_vectorizer, final_tfidf = tfidf(df['Full Text'], df['Category'])
final_tfidf
```

## Deklarasi Fitur X dan Y

Proses ini memisahkan data ke dalam dua kelompok penting: fitur (X) dan target (y) dalam sebuah DataFrame yang disebut df, yang biasanya digunakan dalam proses pembelajaran mesin untuk melakukan prediksi atau klasifikasi. Berikut adalah penjelasan singkatnya:

- **X = df.drop(['Dokumen', 'Label'], axis=1)**: Perintah ini membentuk DataFrame X yang berisi fitur atau atribut dari data. Dalam kasus ini, 'Dokumen' dan 'Label' dihapus dari DataFrame df menggunakan metode `drop` dengan parameter axis=1, yang berarti menghapus kolom. Hasilnya adalah DataFrame X yang terdiri dari semua kolom kecuali 'Dokumen' dan 'Label'. Fitur-fitur ini akan digunakan sebagai masukan untuk pembelajaran mesin.

- **y = df['Label']**: Perintah ini membentuk Series y yang berisi target atau label yang ingin diprediksi. Di sini, hanya kolom 'Label' yang diambil dari DataFrame df dan disimpan dalam Series y. Ini adalah variabel yang akan menjadi target atau label yang ingin diprediksi dalam proses klasifikasi atau prediksi.

Proses ini penting karena mempersiapkan data dengan cara yang diperlukan untuk melatih dan menguji model pembelajaran mesin. Dengan memisahkan fitur (X) dan target (y), kita bisa menggunakan X sebagai input dan y sebagai output yang diinginkan dalam model pembelajaran.

```python
X = final_tfidf.drop(['Dokumen', 'Category'], axis=1)
X
```

```python
y = df['Category']
y
```

## Splitting Data

Pemisahan dataset menjadi subset pelatihan dan pengujian menggunakan fungsi `train_test_split` dari `sklearn.model_selection`.

Ini dilakukan dengan tujuan untuk membagi dataset menjadi empat subset yang berbeda:

- `X_train`: Subset dari fitur (atau variabel independen) yang akan digunakan untuk melatih model.
- `X_test`: Subset lain dari fitur yang akan digunakan untuk menguji model.
- `y_train`: Subset dari target (atau variabel dependen) yang sesuai dengan `X_train`. Ini berisi label atau output yang sesuai dengan subset pelatihan.
- `y_test`: Subset dari target yang sesuai dengan `X_test`. Ini berisi label atau output yang sesuai dengan subset pengujian.

Pemisahan ini dilakukan secara acak dengan `test_size=0.3`, yang berarti 30% dari data akan digunakan sebagai data pengujian, sedangkan 70% yang tersisa akan digunakan sebagai data pelatihan.

Parameter `random_state=42` digunakan untuk menetapkan keadaan awal yang sama saat pemisahan data dilakukan. Ini memastikan bahwa hasil pemisahan data akan konsisten jika kode ini dieksekusi berulang kali dengan nilai `random_state` yang sama.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Modeling LDA

**Mencari Best Parameter**

```python
def find_best_lda(Xtrain, Xtest, n_components, alpha, beta):
  looping = 1
  best = {'k' : 0, 'alpha' : 0, 'beta' : 0, 'accuracy' : 0, 'model': '', 'lda' : '', 'lda_Xtrain' : '', 'lda_Xtest' : ''}
  history = pd.DataFrame(columns=["Pengujian Ke", "K", "Alpha", "Beta", "Accuracy"])

  # Menambahkan tqdm pada loop terluar
  for k in n_components:
    for a in alpha:
      for b in beta:
        lda = LatentDirichletAllocation(n_components=k, doc_topic_prior=a, topic_word_prior=b)
        lda_Xtrain = lda.fit_transform(Xtrain)
        lda_Xtest = lda.transform(Xtest)

        # Membuat model Naive Bayes
        model = GaussianNB()

        # Melatih model pada data pelatihan
        model.fit(lda_Xtrain, y_train)

        # Melakukan prediksi pada data pengujian
        y_pred = model.predict(lda_Xtest)

        # Menghitung akurasi
        accuracy = accuracy_score(y_test, y_pred)
        print(f"Jumlah Topik: {k}, Alpha: {a}, Beta: {b}, Accuracy: {accuracy}")

        if accuracy > best['accuracy']:
          best['accuracy'] = accuracy
          best['k'] = k
          best['alpha'] = a
          best['beta'] = b
          best['model'] = model
          best['lda'] = lda
          best['lda_Xtrain'] = lda_Xtrain
          best['lda_Xtest'] = lda_Xtest

        history.loc[len(history)] = [f"Pengujian Ke- {looping}", k, a, b, accuracy]
        looping += 1

  return (best, history)

k = [3, 4, 5]
alpha = [0.3, 0.4]
beta = [0.1, 0.2]
best_param, history = find_best_lda(X_train, X_test, k, alpha, beta)
```

## Naive Bayes Dengan LDA

```python
# Membuat model Naive Bayes
model = best_param['model']

# Melakukan prediksi pada data pengujian
y_pred = model.predict(lda_x_test)

# Menghitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi:", accuracy)

# Menampilkan laporan klasifikasi
print("Laporan Klasifikasi:")
print(classification_report(y_test, y_pred))

# Menampilkan matriks kebingungan
confusion = confusion_matrix(y_test, y_pred)
print("Confusion Matriks:")
print(confusion)
```
